```python
df['COMPOSITE_DEATH_DATE'] = pd.to_datetime(df['COMPOSITE_DEATH_DATE'], errors='coerce')
```

```python
def map_race(x):
    if x == 1:
        return 'White'
    elif x == 2:
        return 'Black'
    elif x == 4:
        return 'Hispanic'
    else:
        return 'Other'

df['race'] = df['ETHCAT'].apply(map_race)
```

```python
def map_insurance(x):
    if x == 1:
        return 'Private'
    elif x == 2:
        return 'Medicaid'
    elif x == 4:
        return 'Uninsured'
    else:
        return 'Private'

df['insurance3'] = df['PRI_PAYMENT_TCR'].apply(map_insurance)
```

```python
cols_needed = [
    'race','insurance3',
    'INIT_AGE','GENDER','INIT_MELD_PELD_LAB_SCORE',
    'REGION','ABO','HCC_EVER_APPR',
    'DAYSWAIT_CHRON','REM_CD','COMPOSITE_DEATH_DATE'
]

df_small = df[cols_needed].copy()
```

```python
def outcome_func(row):
    rem = row['REM_CD']

    if pd.notnull(rem) and rem == 4:
        return 'Transplant'
    if pd.notnull(row['COMPOSITE_DEATH_DATE']):
        return 'Death'
    return 'Other'

df_small['outcome'] = df_small.apply(outcome_func, axis=1)
```

```python
import numpy as np
import scipy.stats as stats

# Function for p-values
def p_value_numeric(var):
    groups = [df_small[df_small['race']==g][var].dropna() for g in df_small['race'].unique()]
    return stats.f_oneway(*groups)[1]

def p_value_categorical(var):
    tab = pd.crosstab(df_small['race'], df_small[var])
    return stats.chi2_contingency(tab)[1]

# Build Table 1
table1 = pd.DataFrame()

# Age
table1['White'] = [df_small[df_small.race=='White']['INIT_AGE'].mean()]
table1['Black'] = [df_small[df_small.race=='Black']['INIT_AGE'].mean()]
table1['Hispanic'] = [df_small[df_small.race=='Hispanic']['INIT_AGE'].mean()]
table1['Other'] = [df_small[df_small.race=='Other']['INIT_AGE'].mean()]
table1.index = ['Age (mean)']
table1['p-value'] = p_value_numeric('INIT_AGE')

# MELD
table1.loc['MELD (mean)'] = [
    df_small[df_small.race=='White']['INIT_MELD_PELD_LAB_SCORE'].mean(),
    df_small[df_small.race=='Black']['INIT_MELD_PELD_LAB_SCORE'].mean(),
    df_small[df_small.race=='Hispanic']['INIT_MELD_PELD_LAB_SCORE'].mean(),
    df_small[df_small.race=='Other']['INIT_MELD_PELD_LAB_SCORE'].mean(),
    p_value_numeric('INIT_MELD_PELD_LAB_SCORE')
]

# Gender (male %)
table1.loc['% Male'] = [
    (df_small[df_small.race=='White']['GENDER'].eq('M').mean()*100),
    (df_small[df_small.race=='Black']['GENDER'].eq('M').mean()*100),
    (df_small[df_small.race=='Hispanic']['GENDER'].eq('M').mean()*100),
    (df_small[df_small.race=='Other']['GENDER'].eq('M').mean()*100),
    p_value_categorical('GENDER')
]

# Median Wait Time
table1.loc['Median Wait (days)'] = [
    df_small[df_small.race=='White']['time'].median(),
    df_small[df_small.race=='Black']['time'].median(),
    df_small[df_small.race=='Hispanic']['time'].median(),
    df_small[df_small.race=='Other']['time'].median(),
    p_value_numeric('time')
]

table1
```
